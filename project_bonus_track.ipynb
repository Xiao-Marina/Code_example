```python
class biblio(object):
 
    def __init__(self, latitude, longitude, name,schedule = [[9,18],[9,18],[9,18],[9,18],[9,18],[9,18],[9,18]]):
        """Constructor"""
        self.latitude = latitude
        self.longitude = longitude
        self.name = name
        l = list()
        n = len(schedule)//2
        for i in range(n):
          l.append([int(schedule[2*i]),int(schedule[2*i+1])])
        self.schedule = l
    
    def timetable(self):
        im = Image.new('RGB', (200,200), color=('#FFFFFF'))
        

        font = ImageFont.truetype('Nevduplenysh-Regular.ttf', size=18)
        dayweek = ["Пн","Вт","Ср","Чт","Пт","Сб","Вс"]
        draw_text = ImageDraw.Draw(im)
        draw_text.text(
            (20, 20),
            'Режим работы',
            font=font,
            fill='#1C0606')
        for i in range(7):
          s = ''
          fill_f = '#000000'
          if self.schedule[i][0] == 0 or self.schedule[i][1] == 0:
            s = dayweek[i]+' Выходной'
            fill_f = '#FF0000'
          elif self.schedule[i][0] < 10:
            s = dayweek[i]+' 0'+str(self.schedule[i][0])+':00 - '+str(self.schedule[i][1])+':00'
          else:
            s = dayweek[i]+' '+str(self.schedule[i][0])+':00 - '+str(self.schedule[i][1])+':00'
          if i > 4:
            fill_f = '#32CD32'
          draw_text.text(
            (40, 40+i*22),
            s,
            font=font,
            fill=fill_f)
        return im
    def timetable_text(self):
      dayweek = ["Пн","Вт","Ср","Чт","Пт","Сб","Вс"]
      s = "Режим работы\n"
      for i in range(7):
        s += dayweek[i]+" " +str(self.schedule[i][0])+":00 - "+str(self.schedule[i][1])+":00\n"
      return s

    def timetable_html(self):
      dayweek = ['''Пн''','''Вт''','''Ср''','''Чт''','''Пт''','''Сб''','''Вс''']
      s = f'''<font size = "4" face="Comic Sans" color="#0000FF"> {self.name} </font> <br>'''
      s += '''<font color="#000000"> Режим работы </font> <br>'''
      for i in range(7):
        fill_f = '''000000'''
        if i > 4:
          fill_f = '''32CD32'''
        if self.schedule[i][0] == 0 or self.schedule[i][1] == 0:
          fill_f = '''FF0000'''
          s +=f'''<font color=#{fill_f}>''' + dayweek[i]+''' Выходной </font><br>'''
          
        elif self.schedule[i][0] < 10:
          s +=f'''<font color=#{fill_f}>'''+dayweek[i]+''' 0'''+str(self.schedule[i][0])+''':00 - '''+str(self.schedule[i][1])+''':00</font><br>'''
        else:
          s += f'''<font color=#{fill_f}>'''+dayweek[i]+''' '''+str(self.schedule[i][0])+''':00 - '''+str(self.schedule[i][1])+''':00</font><br>'''
        

      return s

    def get_coord(self):
      return [self.latitude,self.longitude]
    
    def show(self):
      print(self.latitude,self.longitude,self.name,self.schedule,sep =', ')
```

```python
def get_biblio(lst):
  latitude = lst[0]
  longitude = lst[1]
  name = lst[2]
  schedule = lst[3:]
  return biblio(latitude,longitude,name,schedule)
```

```python
def make_route(start_point,end_point):
    parameters = {
        'api_key': API_KEY,
        'start' : '{},{}'.format(start_point[1], start_point[0]),
        'end' : '{},{}'.format(end_point[1], end_point[0])
    }
    response = requests.get(
        'https://api.openrouteservice.org/v2/directions/foot-walking', params=parameters)

    if response.status_code == 200:
        data = response.json()
    else:
        print('Request failed:',response.status_code)
        return (0,-1)
    route= data['features'][0]['geometry']['coordinates']
    summary = data['features'][0]['properties']['summary']
    route_xy = [(y, x) for x, y in route]
    distance = summary['distance']
    return (route_xy,distance)
```

```python
def draw_route(start_point,bibl):
    m_temp = folium.Map(location=[59.949178,30.316445], zoom_start=10)
    start_name = "Ваше положение"
    iframe = folium.IFrame(bibl.timetable_html())
    popup = folium.Popup( iframe,
                     min_width=200,
                     max_width=500)
    end_point = bibl.get_coord()
    route_xy,distance = make_route(start_point,end_point)
    folium.Marker(end_point, popup=popup, 
                  icon=folium.Icon(color='green', icon='crosshairs', prefix='fa')
                 ).add_to(m_temp)
    folium.Marker(start_point, popup=start_name,
                  icon=folium.Icon(
                      color='blue', icon='crosshairs', prefix='fa')
                 ).add_to(m_temp)
    if distance == -1:
        print('Произошли технические шоколадки')
        return -1


    folium.PolyLine(route_xy, tooltip='Расстояние: {}м'.format(distance)).add_to(m_temp) 
    return m_temp
```
